```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
df = pd.read_csv(r'D:\Umang\AI & ML\Predictive Maintenance\predictive_maintenance.csv')
df.head()
```

```python
df['Air temperature [K]'] = df['Air temperature [K]'] - 272.15
df['Process temperature [K]'] = df['Process temperature [K]'] - 272.15
```

```python
df.pop('UDI')
df.pop('Product ID')
df.head()
```

```python
df.rename(columns={'Air temperature [K]':'Air temperature [C]', 'Process temperature [K]':'Process temperature [C]'}, inplace=True)
df.head()
```

```python
df = pd.get_dummies(df, columns=['Type'])
df[['Type_H','Type_L','Type_M']] = df[['Type_H','Type_L','Type_M']].replace({False:0, True:1})
df.head()
```

```python
X = df.drop(['Target','Failure Type'], axis=1)
y = df['Failure Type']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, shuffle=True)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
rfc = RandomForestClassifier()
rfc.fit(X_train, y_train)
rfc_pred = rfc.predict(X_test)
rfc_score = round(rfc.score(X_train, y_train)*100,2)
rfc_accu = accuracy_score(rfc_pred, y_test)*100
print("The training accuracy is ", rfc_score,"%")
print("The model accuracy is ", rfc_accu,"%")
```

```python
print("Random Forest Classifier")
print(classification_report(rfc_pred, y_test))
print("Confusion matrix: ") 
confusion_matrix(rfc_pred, y_test)
```

```python
Data = {'Ground Truth':y_test, 'Predicted Failure':rfc_pred}
new = pd.DataFrame(Data)
new.to_csv(r'D:\Umang\AI & ML\Predictive Maintenance\submitted1.csv',index=False)
```

```python
def make_prediction():
    air_temp = input('Enter the Air Temperature[C]: ')
    pro_temp = input('Enter the Process Temperature[C]: ')
    Rot_speed = input('Enter the Rotational Speed[rpm]: ')
    tor = input('Enter the Torque[Nm]: ')
    tool_wear = input('Enter the Tool wear[min]: ')
    Type = input('Enter the Type of machine(L/M/H): ')
    match Type:
        case 'L':
            Type_H = 0
            Type_L = 1
            Type_M = 0
        case 'M':
            Type_H = 0
            Type_L = 0
            Type_M = 1
        case 'H':
            Type_H = 1
            Type_L = 0
            Type_M = 0
    user_inp = [air_temp,pro_temp,Rot_speed,tor,tool_wear,Type_H,Type_L,Type_M]
    X_user = pd.DataFrame([user_inp])
    X_user.columns = X_test.columns
    result = rfc.predict(X_user)[0]
    print(result)
```

```python
choice = input('Is there any input(Y/N)?')
if choice == 'Y':
    make_prediction()
```
